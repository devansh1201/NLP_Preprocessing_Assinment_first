```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
#uploding text data
```

```python
text_file="Chess is a strategic board game played between two players.Each player begins with sixteen pieces: one king, one queen, two rooks, two bishops, two knights, and eight pawns.he main objective is simple: checkmate the opponent’s king!Players must think carefully before every move, because one mistake can change the entire game.Chess improves concentration, patience, and problem-solving skills."
```

```python
print(text_file)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
# punctuation removing
tokens = word_tokenize(text_file)
tokens_punctuation=[word for word in tokens if word not in string.punctuation]
print("\nafter removing punctuation:-")
print(tokens_punctuation)
```

```python
import nltk
nltk.download('stopwords')
```

```python
#stop words removal
stop_words = set(stopwords.words('english'))
filtered_words=[w for w in tokens_punctuation if w.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_words=[stemmer.stem(word) for word in filtered_words]

print(stemmed_words)
```

```python
#Lemmatization

import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in filtered_words]

print(lemmatized_words)
```
